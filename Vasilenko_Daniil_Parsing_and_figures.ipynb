```python

from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support import expected_conditions as ec
from selenium.webdriver.support.ui import WebDriverWait
from selenium.common.exceptions import ElementNotInteractableException
from selenium.webdriver.common.keys import Keys
import time

service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service) 
driver.get("https://bubble-home.ru/catalog")

basket_info = []
try:  
    while True:
        try:
            driver.find_element(By.TAG_NAME, 'body').send_keys(Keys.END)
            time.sleep(2)
            product_containers = WebDriverWait(driver, 10).until(
                ec.presence_of_all_elements_located(
                    (By.CSS_SELECTOR, ".js-store-prod-name.js-product-name.t-store__card__title.t-typography__title.t-name.t-name_xs, .js-product-price.js-store-prod-price-val.t-store__card__price-value")))
            for element in product_containers:
                text = element.text
                print(text)
                basket_info.append(text)
            time.sleep(2)
            driver.execute_script("window.scrollBy(0, -100);")
            driver.find_element(By.XPATH, "//td[text()='Загрузить еще']").click()
        except ElementNotInteractableException:
            break
finally:
    driver.quit()
   
juice_balls_prices = [basket_info[n+1] for n, x in enumerate(basket_info) if "Джус" in x]
juice_balls_prices1 = [int(int(x)*0.75) for x in juice_balls_prices]
print(juice_balls_prices1)

juice_balls = [text for text in basket_info if "Джус" in text]
  
name_juice_balls = [text.split()[3:] for text in juice_balls]

#гпт
name_juice_balls1 = [' '.join(words) if len(words) > 1 else words[0] for words in name_juice_balls if words]
print(name_juice_balls1)

print(basket_info)

#подгатавливаем список, который понадобится позже
list_names_and_prices1 = []
for a, b in zip(name_juice_balls1, juice_balls_prices1):
    list_names_and_prices1.append(a)
    list_names_and_prices1.append(b)

print(list_names_and_prices1)
```

```python

service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service) 
driver.get("https://bubbleteamarket.ru/catalog/toppings/")
basket_info = []

try:
    driver.find_element(By.TAG_NAME, 'body').send_keys(Keys.END)
    time.sleep(2)
    product_containers = WebDriverWait(driver, 10).until(ec.presence_of_all_elements_located((By.CSS_SELECTOR, ".item-title, .price_value")))
    for element in product_containers:
        text = element.text
        print(text)
        basket_info.append(text)
     
finally:
    driver.quit()  

#получаем цены
juice_balls_prices = ["".join(basket_info[n+1].split()) for n, x in enumerate(basket_info) if "Джус" in x]
juice_balls_prices5 = [int(int(x[:5])//3) for x in juice_balls_prices]
print(juice_balls_prices5)

#получаем виды джус боллов
name_juice_balls5 = [text.split('—')[-1].strip().lower() for text in basket_info if "Джус" in text]
print(name_juice_balls5)


#подгатавливаем список, который понадобится позже
list_names_and_prices5 = []
for c, d in zip(name_juice_balls5, juice_balls_prices5):
    list_names_and_prices5.append(c)
    list_names_and_prices5.append(d)

print(list_names_and_prices5)
```

```python
#формирую датасет для графика
import pandas as pd
names_all_balls = name_juice_balls1 + name_juice_balls3 + name_juice_balls5
names_all_balls = list(set(names_all_balls))
bubble_home = name_juice_balls1
bubble_team = name_juice_balls3
bubble_tea = name_juice_balls5


lists = {
    "bubble_home": set(bubble_home),
    "bubble_team": set(bubble_team),
    "bubble_tea": set(bubble_tea)
}
 
data = {}
for list_name, elements in lists.items():
    #с этой строчкой гпт помог
    data[list_name] = [1 if item in elements else 0 for item in names_all_balls]
    
df = pd.DataFrame(data, index=names_all_balls).T

df
```

```python
#спрашивал у гпт про то, как цифры подписывать, обвести чёрным ячейки и тому подобное
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6)) 

heatmap = sns.heatmap(data=df, cmap=['white', '#1f77b4'], annot=True, linewidths=0.5, linecolor='black', cbar=False,)

plt.title('Наличие вкусов у конкурентов') 
plt.ylabel('Конкуренты', labelpad=10) 

heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=45, horizontalalignment='right')

plt.tight_layout() 
plt.show()
```

```python
#получаем словари для, где товар: цена для каждого магазина. Немножео помог гпт
bubble_home = {list_names_and_prices1[i]: list_names_and_prices1[i+1] for i in range(0, len(list_names_and_prices1), 2) 
               if i+1 < len(list_names_and_prices1)}
bubble_team = {list_names_and_prices3[i]: list_names_and_prices3[i+1] for i in range(0, len(list_names_and_prices3), 2) 
               if i+1 < len(list_names_and_prices3)}
bubble_tea = {list_names_and_prices5[i]: list_names_and_prices5[i+1] for i in range(0, len(list_names_and_prices5), 2) 
               if i+1 < len(list_names_and_prices5)}

full_shops = {'bubble_home': bubble_home,
              'bubble_team': bubble_team,
              'bubble_tea': bubble_tea}
print(full_shops)

list_df = []

for store_name, products in full_shops.items():
    df1 = pd.DataFrame(list(products.items()), columns=['product', 'price'])
    df1['store'] = store_name
    list_df.append(df1)

df_long = pd.concat(list_df, ignore_index=True)

df_long = df_long[['store', 'product', 'price']]

df_long = df_long.drop(15)

#теперь строим график
plt.figure(figsize =(14, 8))
barplot = sns.barplot(data = df_long, x = 'product', y = 'price', hue = 'store', palette = 'viridis', gap = 0.3)
barplot.set_xticklabels(barplot.get_xticklabels(), rotation=45, horizontalalignment='right')

plt.tight_layout() 
plt.show()

df_long
```
