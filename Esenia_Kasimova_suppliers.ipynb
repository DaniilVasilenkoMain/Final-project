```python
#Спарсили цены на топиоку с алибабы
import pandas as pd 
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support import expected_conditions as EC
import time
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support.ui import WebDriverWait
service = Service(ChromeDriverManager().install())
dr = webdriver.Chrome(service=service) 
dr.get("https://www.alibaba.com/trade/search?exportCountry=RU&keywords=Tapioca")       
dr.execute_script("window.scrollTo(0, document.body.scrollHeight);")
time.sleep(3)
wait = WebDriverWait(dr, 7)
a = wait.until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, ".search-card-e-title, .search-card-e-price-main, .search-card-e-company.margin-bottom-12")))
tripple_list = []
for b in a:
    words = b.text
    tripple_list.append(words)       
data = []
for n in range(0, len(tripple_list), 3):
        product = tripple_list[n]
        price = tripple_list[n+1]
        maker = tripple_list[n+2]
        data.append({'Maker': maker, 'Product': product, 'Price': price})
df = pd.DataFrame(data)

dr.quit()

df_copy = df.copy()

df_copy = df[df.apply(lambda row: row.astype(str).str.contains('тапиок', case=False, na=False).any(),axis=1)] #подсказал гпт. Оставили только тапиоку

#чтобы получить цены как int и работать с ними дальше
def nice_last_column(price):
    price = price.replace('$', '').replace(' ', '').strip()  
    price = price.replace(',', '.')
    split_price = price.split('-')
    nums = [float(part) for part in split_price]
    averege = sum(nums) / len(nums)
    return float(round(averege,2)*85)

df_copy['Price'] = df['Price'].apply(nice_last_column)

df_copy = df_copy.sort_values(by=df_copy.columns[-1], ascending=True)
df_copy.columns = ['Maker', 'Product', 'Price ₽']
#сверху самая маленькая цена 
df_copy
```
